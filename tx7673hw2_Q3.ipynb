```python
traindf = pd.read_csv(r"/Users/STSC/Desktop/cs697a/hw2/optdigits_69_N200.tra",header=None)
train_df = pd.read_csv(r"/Users/STSC/Desktop/cs697a/hw2/optdigits_69_N200.tra",header=None)
testdf = pd.read_csv(r'/Users/STSC/Desktop/cs697a/hw2/optdigits_69.tes',header=None)
test_df = pd.read_csv(r'/Users/STSC/Desktop/cs697a/hw2/optdigits_69.tes',header=None)
train_df = traindf.drop(traindf.columns[64],axis=1)
test_df = testdf.drop(testdf.columns[64],axis=1)
```

```python
#Q1
class6 = train_df[:100]
test6 = test_df[:180]
class9 = train_df[100:]
test9 = test_df[180:]
testerrors = {}
trainerrors = {}
class6mean = []
class9mean = []
class6std = []
class9std = []
m = 64
for i in range(m):
    temp1 = np.std(class6[i])
    class6std.append(temp1)
    temp2 = np.std(class9[i])
    class9std.append(temp2)
    temp3 = np.mean(class6[i])
    class6mean.append(temp3)
    temp4 = np.mean(class9[i])
    class9mean.append(temp4)
```

```python
best = {}
for col in range(m):
    errors = 0
    for i in range(200):
        if (class6std[col] != 0):
             g6 = -(np.log(2*np.pi)/2) - np.log(class6std[col]) - ((traindf[col][i]-class6mean[col])**2)/(2*((class6std[col])**2)) + np.log(0.5)
        else:
            continue

        if (class9std[col] != 0):
            g9 = -(np.log(2*np.pi)/2) - np.log(class9std[col]) - ((traindf[col][i]-class9mean[col])**2)/(2*((class9std[col])**2)) + np.log(0.5)
        else:
            continue
        if g6 < g9:
            if i < 100:
                errors += 1
        else:
            if i >= 100:
                errors += 1
                
    if errors != 0:
        trainerrors[col] = errors
print("Training errors: ", trainerrors)
print("\n")

for col in range(m):
    errors = 0
    for i in range(361):
        if (class6std[col] != 0):
             t6 = -(np.log(2*np.pi)/2) - np.log(class6std[col]) - ((testdf[col][i]-class6mean[col])**2)/(2*((class6std[col])**2)) + np.log(0.5)
        else:
            continue

        if (class9std[col] != 0):
            t9 = -(np.log(2*np.pi)/2) - np.log(class9std[col]) - ((testdf[col][i]-class9mean[col])**2)/(2*((class9std[col])**2)) + np.log(0.5)
        else:
            continue
        if t6 < t9:
            if i < 180:
                errors += 1
        else:
            if i >= 180:
                errors += 1
                
    if errors != 0:
        testerrors[col] = errors
print("Test errors: ", testerrors)
print("\n")
best = sorted(testerrors.items(), key=lambda k:k[1])
best = best[:10]
print("Best 10 features are: ", best)
```
