```python
# Classified the dataset as per class label and summarizing mean and variance

my_dataset_summary = my_training_dataset_data.to_numpy()
misclassified_value = 0

def classify_by_class(dataset_loc):
    classified = dict()
    for k in range(len(dataset_loc)):
        vector_loc = dataset_loc[k]
        class_value_loc = vector_loc[-1]
        if (class_value_loc not in classified):
            classified[class_value_loc] = list()
        classified[class_value_loc].append(vector_loc)
    return classified    
 
def summarize_dataset_data(dataset_loc):
    summaries_result = [(npy.mean(column_val), npy.var(column_val)) for column_val in zip(*dataset_loc)]
    del(summaries_result[-1])
    return summaries_result
```

```python
# calculating mean, variance of 6 and 9 and putting them in different list

summary_for_6_trainingdata = list(summarize_dataset_data(classify_by_class(my_dataset_summary).get(6)))
summary_for_9_trainingdata = list(summarize_dataset_data(classify_by_class(my_dataset_summary).get(9)))
```

```python
my_classif = my_training_dataset_data[64]

errorlist_training= list()

# Here I have calculated the g_of_6 and g_of_9 for each instance of non-zero variance columns 
for column_val in my_training_dataset_data:
    misclassified_value = 0
    if column_val == 64:
        break
        
    mean_for_6 = summary_for_6_trainingdata[column_val][0]
    variance_for_6 = summary_for_6_trainingdata[column_val][1]
    mean_for_9 = summary_for_9_trainingdata[column_val][0]
    variance_for_9 = summary_for_9_trainingdata[column_val][1]
    
    if variance_for_6 == 0 or variance_for_9 == 0:
        prediction_is = "It can't be classified"
        error_is = "It can't be classified"
        errorlist_training.append(error_is)
        continue
    # Here I have predicted the class and found the errors.
    for row_val in range(0, len(my_training_dataset_data)):
        vector_loc = my_training_dataset_data[column_val][row_val]
        g_of_6_x = -(0.5 * math.log10(2 * math.pi))
        g_of_6_x = g_of_6_x - math.log10(variance_for_6)
        g_of_6_x = g_of_6_x - (math.pow((vector_loc - mean_for_6), 2) / 2 * math.pow(variance_for_6, 2)) + math.log10(0.5)
        g_of_9_x = -(0.5 * math.log10(2 * math.pi))
        g_of_9_x = g_of_9_x - math.log10(variance_for_9)
        g_of_9_x = g_of_9_x - (math.pow((vector_loc - mean_for_9), 2) / 2 * math.pow(variance_for_9, 2)) + math.log10(0.5)
        print("%7s %7s %7s %7s" % (row_val, column_val, g_of_6_x, g_of_9_x))
        if(g_of_6_x > g_of_9_x ):
            prediction_is = 6
        else:
            prediction_is = 9
           
        if (prediction_is != my_classif[row_val]):
            misclassified_value = misclassified_value + 1
   
    error_is = misclassified_value/361
    errorlist_training.append(error_is)
```
